```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
food = pd.read_csv('../data/seasonality-values-d2b9a3.csv')
```

```python
food.columns
```

```python
food.week_id.dtype
```

```python
food['week_num'] = food.week_id.str.split('-', expand=True)[1].astype(int)
```

```python
food['year'] = food.week_id.str.split('-', expand=True)[0]
food['year'] = pd.to_datetime(food.year)
```

```python
food.head()
```

```python
food.id.unique()
```

```python
import datetime
```

```python
food['date'] = food.year + datetime.timedelta(days=7) * food.week_num
```

```python
berry = pd.DataFrame({'weekly':food[food.id=='blackberry'].set_index('date')['value'], 'roll_4':food[food.id=='blackberry'].set_index('date')['value'].rolling(4).mean(), 'avg':food[food.id=='blackberry'].set_index('date')['value'].mean(), 'expand': food[food.id=='blackberry'].set_index('date')['value'].expanding().mean()})
```

```python
foods = [berry, garlic, pie, turkey, honey, empanada, chia, pho, coffee]
foods_names = ['berry', 'garlic', 'pie', 'turkey', 'honey', 'empanada', 'chia', 'pho', 'coffee']
```

```python
fig, axs = plt.subplots(len(foods), 1, figsize=(18,24))

for idx, ax in enumerate(axs.flatten()):
    ax.bar(foods[idx][:200].index, foods[idx][:200]['weekly'])
    ax.plot(foods[idx][:200][['roll_4', 'avg', 'expand']])
    ax.set_title(f'{foods_names[idx]}')
fig.tight_layout()
```
